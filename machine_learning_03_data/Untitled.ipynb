```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("test_sample.csv")
```

```python
data
```

```python
X, y = data.drop("Y", axis=1), data["Y"]
```

```python
grid = {"ccp_alpha": np.linspace(0.01,0.1,10)}
regr_tree = DecisionTreeRegressor(random_state=0)
cross_val = GridSearchCV(regr_tree, grid, cv=10, scoring="neg_mean_squared_error")
cross_val.fit(X, y)
print(cross_val.best_estimator_)
print(-cross_val.best_score_)
```
