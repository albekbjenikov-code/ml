```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("test_sample.csv")
```

```python
df
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
lm0 = sm.OLS(y, sm.add_constant(X)).fit()
lm0.summary()
```

```python
eliminated_by_lm = [int(s[1:]) for s in lm0.pvalues[lm0.pvalues >= 0.05].index]
```

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_mod = LassoCV(cv=5, random_state=1)
lasso_mod.fit(X, y)
eliminated_by_Lasso = [int(s[1:]) for s in X.columns[lasso_mod.coef_ == 0]]
```

```python
eliminated_by_lm
```

```python
lasso_zeros = ' '.join([str(idx) for idx in eliminated_by_Lasso])
lm_zeros = ' '.join([str(idx) for idx in eliminated_by_lm])
pd.DataFrame([lasso_zeros,lm_zeros], index = ['eliminated_by_Lasso','eliminated_by_lm']).to_csv('answer.csv')
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.linear_model import LassoCV
import statsmodels.api as sm
data=df

# Ð Ð°Ð·Ð´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð½Ð° Ð¿Ñ€ÐµÐ´Ð¸ÐºÑ‚Ð¾Ñ€Ñ‹ Ð¸ Ñ†ÐµÐ»ÐµÐ²ÑƒÑŽ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½ÑƒÑŽ
X = data.drop('Y', axis=1)
y = data['Y']

# 1. LASSO Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ñ
lasso_cv = LassoCV(cv=5, random_state=1)
lasso_cv.fit(X, y)

# ÐÐ°Ñ…Ð¾Ð´Ð¸Ð¼ Ð¸Ð½Ð´ÐµÐºÑÑ‹ Ð¸ÑÐºÐ»ÑŽÑ‡ÐµÐ½Ð½Ñ‹Ñ… Ð¿Ñ€ÐµÐ´Ð¸ÐºÑ‚Ð¾Ñ€Ð¾Ð² (ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ñ‹ = 0)
lasso_coef = lasso_cv.coef_
eliminated_by_Lasso = np.where(lasso_coef == 0)[0].tolist()

# 2. ÐžÐ±Ñ‹Ñ‡Ð½Ð°Ñ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð°Ñ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ñ
# Ð”Ð¾Ð±Ð°Ð²Ð»ÑÐµÐ¼ ÐºÐ¾Ð½ÑÑ‚Ð°Ð½Ñ‚Ñƒ Ð´Ð»Ñ intercept
X_with_const = sm.add_constant(X)

# Ð¡Ñ‚Ñ€Ð¾Ð¸Ð¼ OLS Ð¼Ð¾Ð´ÐµÐ»ÑŒ
model = sm.OLS(y, X_with_const).fit()

# ÐŸÐ¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ p-values (Ð¸ÑÐºÐ»ÑŽÑ‡Ð°Ñ intercept)
p_values = model.pvalues[1:]  # Ð¿Ñ€Ð¾Ð¿ÑƒÑÐºÐ°ÐµÐ¼ intercept (Ð¿ÐµÑ€Ð²Ñ‹Ð¹ ÑÐ»ÐµÐ¼ÐµÐ½Ñ‚)

# ÐÐ°Ñ…Ð¾Ð´Ð¸Ð¼ Ð¸Ð½Ð´ÐµÐºÑÑ‹ Ð¿Ñ€ÐµÐ´Ð¸ÐºÑ‚Ð¾Ñ€Ð¾Ð² Ñ p-value > 0.1
eliminated_by_lm = np.where(p_values > 0.1)[0].tolist()

# ÐŸÐ¾Ð´Ð³Ð¾Ñ‚Ð¾Ð²ÐºÐ° Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð° Ð´Ð»Ñ ÑÐ¾Ñ…Ñ€Ð°Ð½ÐµÐ½Ð¸Ñ
lasso_zeros = ' '.join([str(idx) for idx in eliminated_by_Lasso])
lm_zeros = ' '.join([str(idx) for idx in eliminated_by_lm])

# Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ Ð¸ ÑÐ¾Ñ…Ñ€Ð°Ð½ÐµÐ½Ð¸Ðµ Ñ„Ð°Ð¹Ð»Ð° Ð¾Ñ‚Ð²ÐµÑ‚Ð°
result_df = pd.DataFrame([lasso_zeros, lm_zeros], 
                        index=['eliminated_by_Lasso', 'eliminated_by_lm'])
result_df.to_csv('answer.csv')

# Ð’Ñ‹Ð²Ð¾Ð´ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð² Ð´Ð»Ñ Ð¿Ñ€Ð¾Ð²ÐµÑ€ÐºÐ¸
print("Ð˜ÑÐºÐ»ÑŽÑ‡ÐµÐ½Ð¾ LASSO:", eliminated_by_Lasso)
print("Ð˜ÑÐºÐ»ÑŽÑ‡ÐµÐ½Ð¾ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð¹ Ð¼Ð¾Ð´ÐµÐ»ÑŒÑŽ:", eliminated_by_lm)
print(f"LASSO Ð¸ÑÐºÐ»ÑŽÑ‡Ð¸Ð» {len(eliminated_by_Lasso)} Ð¿Ñ€ÐµÐ´Ð¸ÐºÑ‚Ð¾Ñ€Ð¾Ð²")
print(f"Ð›Ð¸Ð½ÐµÐ¹Ð½Ð°Ñ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð¸ÑÐºÐ»ÑŽÑ‡Ð¸Ð»Ð° {len(eliminated_by_lm)} Ð¿Ñ€ÐµÐ´Ð¸ÐºÑ‚Ð¾Ñ€Ð¾Ð²")
```
